```python
from datasets import load_dataset
import json
```

```python
with open("./mathvision_mistake_correction_v2.json", "r") as f:
    file = json.load(f)
```

```python

cleaned_dict= {}
for item in file:
    conversations = item["conversations"]
    row_arr = []
    for text_dic in conversations:
        try:
            row_arr.append({"student": text_dic["student"]})
        except:
            pass

        try:
            row_arr.append({"teacher": text_dic["teacher"]})
        except:
            pass

    cleaned_dict[item["id"]] = {"mistake_correction_conv": row_arr}

cleaned_dict
```

```python
with open("mathvision_mistake_correction_v3.json", "w") as f:
    json.dump(cleaned_dict, f)
```
